**Step One**: Importing required packages

```python
import pandas as pd;
import numpy as np;
```

<hr/>

**Step Two**: Displaying and analyzing the Dataset characteristics

```python
df = pd.read_csv("Group_21_a.csv", delimiter=",");
display(df, df.info())
```

<div style="font-size:16px"><strong>Step Two Results</strong></div> 

The given dataset suffers from different problems, which are the following:

**- Missing Data**: Columns `Day`, `Greg. Month`, `Location` and `Store Size` have missing values<br/>
**- Type inconsistencies**: Columns `Work Hours`, `Cost/day`, `Profit/ day` should be converted to **`int64`** dtype<br/>
**- Unneeded Data**: Column `0` serves no purpose in the dataset<br/>
**- Non-uniform values**: Columns `Day` and `Greg. Month` contain both full names and shorthand names for their values. For example, `Tue` and `Tuesday`

<hr/>

**Step Three**: The following will be applied to fix dataset flaws

**- Altering Type Inconsistencies**: by removing Strings from the numerical columns<br/>
**- Replacing Missing Data**: replacing missing data by the mean for non-numeric columns & the mode for numeric columns, as        dropping the information of missing values results in a substantial loss in data samples<br/>
**- Dropping Unneeded Data**: To increase the performance of the dataset analysis<br/>
**- Obtaining uniform data**: Changing all values for `Day` and `Greg. Month` to shorthand names;


```python
# Altering Type Inconsistencies
df["Work Hours"] = df["Work Hours"].apply(lambda x: int(x.replace(" hrs", "")))

## all Data has .0 decimal
## Converting to int64 is better for memory and analysis performance
df["Cost/day"] = df["Cost/day"].apply(lambda x: int(float(x.replace("SAR ", "")))) 
df["Profit/day"] = df["Profit/day"].apply(lambda x: int(x.replace("SAR ", "")));




# Replacing Missing Data
null_cols = df.columns[df.isna().any()]
for c in null_cols:
    value = df[c].mean() if df[c].dtype!='object' else df[c].mode()[0]
    df[c].fillna(value,inplace=True)
df.info()

# Dropping Unneeded data
df.drop("0", axis=1, inplace= True)
df

#obtaining uniform data
df["Day"] = df["Day"].apply(lambda x: x[0:3])
df["Greg. Month"] = df["Greg. Month"].apply(lambda x: x[0:3])
```

<hr/>

**Step Four**: Deleting Outliers (Outside the range of whiskers)

```python
## Outliers occur only in numerical data
num_cols = df.select_dtypes(exclude="object").columns;

for col in num_cols:
    Q1, Q3 = np.percentile(df[col].values, [25, 75]);
    IQR = Q3 - Q1;
    topWhisker = Q3 + IQR * 1.5
    bottomWhisker = Q1 - IQR * 1.5
    out_mask = (df[col] > topWhisker) | (df[col] < bottomWhisker);
    df.drop(df.loc[out_mask, col].index, inplace= True)
df
```

**Date Categorical Data**

```python
## Createing a filter to group months in quarters for each calendar type
def hijriQuarters(x):
    x_int =  int(x.split("-")[1])
    yr = x.split("-")[0];
    
    if x_int in range(1, 4):
        return yr + "- Q1"
    if x_int in range(4, 7):
        return yr + "- Q2"
    if x_int in range(7, 10):
        return yr + "- Q3"
    if x_int in range(10, 12):
        return yr + "- Q4"

def gregorianQuarters(x):
    x_int =  int(x.split("/")[1])
    yr = x.split("/")[2].replace(" 0:00", "")
    if x_int in range(1, 4):
        return yr + "- Q1"
    if x_int in range(4, 7):
        return yr + "- Q2"
    if x_int in range(7, 10):
        return yr + "- Q3"
    if x_int in range(10, 12):
        return yr + "- Q4"

df["hijri-quarters"] = df["Hijri Date"].apply(hijriQuarters);
df["gregorian-quarters"] = df["Gregorian Date"]. apply(gregorianQuarters);
display
in_data_dates = ["hijri-quarters", "gregorian-quarters"]
## plotting Fiscal quarters histograms
fig, axes = plt.subplots(1, 2, figsize=(10, 10));
for ind, axis in enumerate(axes.flatten()):
    if(ind == len(in_data_num)):
        break;
    sns.histplot( x = "Profit/day", bins=15, hue= in_data_dates[ind], ax = axis, data = df );
plt.show()

```

## Task-8

**Step One:** importing the needed packages

```python
## Regression modelling packages
from sklearn.linear_model import LinearRegression;
from sklearn.linear_model import RidgeCV;
from sklearn.linear_model import LassoCV;
## Error Calculation Measures
from sklearn.metrics import mean_squared_error;
## Label Encoding Packages
from sklearn.preprocessing import OrdinalEncoder;
from sklearn.preprocessing import StandardScaler;
```

**Step Two**:
<br/>
<br/>
    1- Extracting information needed for training regression models<br/>
    2- Preparing and cleaning testing data<br/>
    2- Label encoding categorical values to utilize regression modeling<br/>
    3- Standard scaling input values for faster computation<br/>

```python
## Exrtacting the information needed for training the regression model
train_df = df.drop(columns = ["Store ID", "Gregorian Date", "Hijri Date"])
test_df = pd.read_csv("Group_21_b.csv").drop(columns = ["Store ID", "Gregorian Date", "Hijri Date"]);
```

```python
## Preparing and cleaning testing data

## Fixing type inconsistencies in testing dataset
test_df["Profit/day"] = test_df["Profit/day"].apply(lambda x: int(x.replace("SAR ", "")))
test_df["Cost/day"] = test_df["Cost/day"].apply(lambda x: int(float(x.replace("SAR ", ""))))
test_df["Work Hours"] = test_df["Work Hours"].apply(lambda x: int(x.replace(" hrs", "")))
 ## Unneeded data
test_df.drop(columns = "0", inplace = True);
```

```python
## encoding the categorical variables
cat_cols = train_df.select_dtypes(include="object").columns;
encoder = OrdinalEncoder()
encoder.fit(train_df.loc[:, cat_cols]);
display(train_df, test_df)
train_df.loc[:, cat_cols] = encoder.transform(train_df.loc[:, cat_cols]);
test_df.loc[:, cat_cols] = encoder.transform(test_df.loc[:, cat_cols]);

display(train_df, test_df)
```

```python
## Standard Scaling all the values;
x_train = train_df.drop(columns=["Profit/day"], axis = 1).values;
y_train = train_df["Profit/day"].values;
x_test = test_df.drop(columns=["Profit/day"], axis=1).values;
y_test = test_df["Profit/day"].values;
y_test_org = y_test;
scaler = StandardScaler();

scaler.fit(np.c_[x_train, y_train]);
A_train = scaler.transform(np.c_[x_train, y_train])
x_train = A_train[:, :-1];
y_train = A_train[:,-1];
A_test = scaler.transform(np.c_[x_test, y_test]);
x_test = A_test[:, :-1];
y_test = A_test[:, -1]
```

**Step Three**: Fit our data in the multiple regression models, to find the most efficient one

```python
## OLS
reg1 = LinearRegression(fit_intercept=False).fit(x_train, y_train);
y_pred1 = reg1.predict(x_test);
# Ridge
reg2 = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3], fit_intercept=False,cv=10).fit(x_train, y_train);
y_pred2 = reg2.predict(x_test);
# Lasso
reg3 = LassoCV(alphas=[1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3],
               fit_intercept=False,cv=10, random_state=0).fit(x_train, y_train)
y_pred3 = reg3.predict(x_test);

# Computing the error of each model to use the most efficent
ols_mse = mean_squared_error(y_test, y_pred1);
ridge_mse = mean_squared_error(y_test, y_pred2);
lasso_mse = mean_squared_error(y_test, y_pred3)
print(f"""
MSE using OLS: {ols_mse}
MSE using Ridge: {ridge_mse}
MSE using Lasso: {lasso_mse}
""")
```

**Step 4:** Using the most efficient model, which is in our case is the lasso penalized regression to make predictions about the daily profit
